```python
#task 7
import os
import glob
import shutil

cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

#to remove Cairo 
shutil.rmtree('/Users/justintzn/Desktop/learning-portfolio-2024-aug-Justintzn/os-collaboration-exercise-data/Cairo')

#putting back Cairo
os.mkdir('/Users/justintzn/Desktop/learning-portfolio-2024-aug-Justintzn/os-collaboration-exercise-data/Cairo')

#moving them in

os.chdir('/Users/justintzn/Desktop/learning-portfolio-2024-aug-Justintzn/os-collaboration-exercise-data')

name = []
for i in range(len(cities)):
    name.append(cities[i].lower())

dic = dict(zip(name,cities))

al = glob.glob('*')

for i in range(len(al)):
    for j in range(len(dic)):
        if list(dic.keys())[j] in al[i]:
            shutil.move(al[i],list(dic.values())[j])


#go through every file
#check for name 
#if yes, move into cities
    
```
